### Balance de clases

Se analizó la distribución de la variable objetivo **cancelacion** para evaluar la existencia de desbalance de clases.

Los resultados muestran que aproximadamente el **74% de los clientes permanecen activos**, mientras que el **26% cancela el servicio**.

Este nivel de desbalance se considera moderado y no representa un problema crítico para el entrenamiento de los modelos. Por esta razón, se decidió entrenar los modelos con el dataset original, sin aplicar técnicas de balanceo como oversampling o SMOTE.

```python
print("===== VALIDACIÓN DEL DATASET =====")

# 1️⃣ Verificar valores únicos de la variable objetivo
print("\nValores de cancelacion:")
print(df['cancelacion'].value_counts())

# 2️⃣ Verificar proporción de clases
print("\nProporción de clases:")
print(df['cancelacion'].value_counts(normalize=True))

# 3️⃣ Verificar valores nulos
print("\nValores nulos:")
print(df.isnull().sum())

# 4️⃣ Verificar tipos de datos
print("\nTipos de datos:")
print(df.dtypes)

# 5️⃣ Verificar dimensiones del dataset
print("\nDimensión del dataset:")
print(df.shape)
```

```python
df.nunique().sort_values()
```

```python
#estableciendo la variable objetivo
X = df.drop('cancelacion', axis=1)
y = df['cancelacion']
```

```python
#borrando la columna id_cliente
X = X.drop('id_cliente', axis=1)
```

```python
#validando que se elimino la columna correctamente de la base de datos
print('id_cliente' in X.columns)
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
# Identificando columnas categóricas
# Reconstructing a temporary DataFrame to correctly use select_dtypes.
temp_X_df = df.drop('cancelacion', axis=1)
temp_X_df = temp_X_df.drop('id_cliente', axis=1)

columnas_categoricas = temp_X_df.select_dtypes(include=['object']).columns
columnas_numericas = temp_X_df.select_dtypes(exclude=['object']).columns
```

```python
#se crea el encoder
encoder = OneHotEncoder(drop='first', sparse_output=False)
```

```python
#aplicando la transformación
preprocesador = ColumnTransformer(
    transformers=[
        ('cat', encoder, columnas_categoricas)
    ],
    remainder='passthrough'
)
```

```python
#transformando el dataset
X = preprocesador.fit_transform(X)
```

```python
#se verifica el resultado
X[:5]
```

```python
#revisando la cantidad de filas y columnas
X.shape
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.3,
    random_state=42
)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
#entrenando el modelo
from sklearn.linear_model import LogisticRegression

modelo = LogisticRegression(max_iter=1000)

modelo.fit(X_train, y_train)
```

```python
#predicciones
y_pred = modelo.predict(X_test)
```

```python
#evaluación
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```

```python
import pandas as pd


categorical_feature_names = preprocesador.named_transformers_['cat'].get_feature_names_out(columnas_categoricas)

all_feature_names = list(categorical_feature_names) + list(columnas_numericas)

coeficientes = pd.Series(modelo.coef_[0], index=all_feature_names)

coeficientes.sort_values().plot(kind='barh', figsize=(10,8))
plt.title("Variables que influyen en la cancelación de clientes")
plt.show()
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

cm = confusion_matrix(y_test, y_pred)

sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')

plt.title("Matriz de Confusión")
plt.xlabel("Predicción")
plt.ylabel("Valor real")

plt.show()
```
